# Data cleaning pre processing

```python
df_clean = df.copy()
```

```python
df_clean.head()
```

```python
df_clean.drop_duplicates(inplace=True)
```

```python
df_clean.shape
```

```python
df_clean.isnull().sum()
```

```python
df_clean.dtypes
```

```python
df_clean['origin'].value_counts()
```

```python
# one hot encoding
df_clean = pd.get_dummies(df_clean, columns=['origin'], drop_first=True)
```

```python
df_clean.head()
```

```python
print(df_clean.columns)
```

```python
# Exclude the 'name' column and handle NaNs in 'horsepower' before converting to int
df_clean_processed = df_clean.drop('name', axis=1).copy()

# Impute missing values in 'horsepower' with the mean
if 'horsepower' in df_clean_processed.columns:
    df_clean_processed['horsepower'].fillna(df_clean_processed['horsepower'].mean(), inplace=True)

# Convert all remaining columns to int, truncating floats
df_clean = df_clean_processed.astype(int)
```

```python
df_clean.head()
```

# Feature engineering and extraction

```python
df_clean.columns
```

```python
threshold = df_clean['mpg'].median()
df_clean['mpg_class'] = np.where(df_clean['mpg'] >= threshold, 1, 0)
```

```python
df_clean_processed.head()
```

```python
X = df_clean_processed.drop('mpg', axis=1)
y = df_clean['mpg_class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
y_pred
```

```python
y_test
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# Random forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model = rf_model.fit(X_train, y_train)
```

```python
model_rf = RandomForestClassifier()
model_rf.fit(X_train, y_train)
```

```python
y_pred_rf = model_rf.predict(X_test)
```

```python
y_pred_rf
```

```python
y_test
```

```python
accuracy_score(y_test,y_pred_rf)
```

```python
confusion_matrix(y_test, y_pred_rf)
```

```python
print(classification_report(y_test, y_pred_rf))
```
